## Create Line Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,5,6]
y=[4,5,6,7,8]
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plotting')
plt.plot(x,y,color='black',linestyle='dashdot',marker='o',markersize=10,markerfacecolor='red',linewidth='2')
plt.grid(True)
```

## Multiple Plot

```python
x1=[1,2,3,4,5]
x2=[1,0,1,4,9]
y1=[2,3,4,6,7]
y2=[1,4,7,8,2]
plt.figure(figsize=(10,5))

plt.subplot(2,2,1)
plt.plot(x1,y2,color='black',linestyle='dashdot')
plt.title('plot 1')

plt.subplot(2,2,2)
plt.plot(x2,y1,color='blue',linestyle='dashdot')
plt.title('plot 2')

plt.subplot(2,2,3)
plt.plot(x2,y2,color='red',linestyle='dashdot')
plt.title('plot 3')

plt.subplot(2,2,4)
plt.plot(x1,y1,color='black',linestyle='dashdot')
plt.title('plot 4')
```

## Bar Plot

```python
categories=['a','b','c','d']
values=[10,20,30,4]
plt.bar(categories,values,color='purple')
plt.xlabel('Categories')
plt.ylabel('points')
plt.title('Bar plot')
plt.show()

''' Category wise data   Comparison '''
```

## Histograms

```python
data=[1,2,2,2,1,2,3,2,3,3,1,3,5,5,5,2,5,6,7,9]
plt.hist(data,bins=5,color='red',edgecolor='black')
plt.xlabel('data')
plt.ylabel('frequency')
plt.title('Histogram plot')
plt.show()

''' bins =  interval/column  '''
```

## Scatter Plot

```python
x=[1,2,5,4,3]
y=[4,1,6,7,9]
plt.scatter(x,y,color='red',marker='x',cmap='coolwarm',alpha=0.8)
plt.title('Scatter plot')
plt.show()

''' alpha  transparency level     parameter ,alpha=0.8 means 80% visible'''
```

## Pie Chart

```python
tags=['A','B','C','D' ]
icons=[30,40,30,20]
colors=['gold','yellow','Red','Blue']
explode=[0,0,0.3,0]
plt.pie(icons,labels=tags,colors=colors,explode=explode,autopct='%1.1f%%')
plt.show()

##autopct  %  
#exolode       
```
